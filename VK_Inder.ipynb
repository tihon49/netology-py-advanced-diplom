```python
import pandas as pd
```

```python
#создаем дата фрейм
cols = ['id', 'first_name', 'last_name', 'bdate', 'city.title', 'photo_max', 'is_friend',
        'twitter', 'instagram', 'university', 'home_town', 'relation', 'personal.smoking',
        'personal.alcohol', 'personal.religion_id', 'interests', 'music', 'activities',
        'movies','tv','books','games','universities','schools','about','relatives','quotes']
df = pd.read_csv('vk_inder.csv', usecols=cols).set_index('id')
df.head(3)
```

## Отберем только нужные параметры

```python
filter_list = ['bdate', 'city.title', 'home_town', 'personal.smoking', 'personal.alcohol', 
               'interests', 'music', 'activities', 'movies','tv','books','games', 'about']
df2 = df[filter_list]
```

```python
#попробовал найти id тех, у кого в музыук есть слово rap
for i in df2.T.iteritems():
    if 'rock' in str(i[1][6]).lower():
        print(i[0])
```

```python
tihon_df = pd.read_csv('tihon.csv', usecols=['id']+filter_list).set_index('id')
tihon_df
```

```python
# for i in df2.index:
#     df2.set_value(i, 'points', 0)

df2['points'] = 0
```

```python
df2.head(2)
```

```python
for i in tihon_df.T.iteritems():
    tihon_music_list = i[1][6].split(', ')
    tihon_films_list = i[1][8].split(', ')
    tihon_books_list = i[1][10].split(', ')
    tihon_games_list = i[1][11].split(', ')
```

```python
# for i in df2.T.iteritems():
#     if 'rock' in str(i[1][6]).lower():
#         print(i[0])
```

```python
#ищем общие книжки
for book in tihon_books_list:
    book = book.lower()
    for i in df2.T.iteritems():
        if book in str(i[1][10]).lower():
            df2['points'][i[0]] += 1
#             print(df2.T[i[0]])
```

```python
#ищем общую музыку
for music in tihon_music_list:
    music = music.lower()
    for i in df2.T.iteritems():
        if music in str(i[1][6]).lower():
            df2['points'][i[0]] += 1
#             print(df2.T[i[0]])
#             print()
```

```python
#ищем общие фильмы
for film in tihon_films_list:
    film = film.lower()
    for i in df2.T.iteritems():
        if film in str(i[1][8]).lower():
            df2['points'][i[0]] += 1
#             print(df2.T[i[0]])
```

```python
#ищем общие игры
for game in tihon_games_list:
    game = game.lower()
    for i in df2.T.iteritems():
        if game in str(i[1][11]).lower():
            df2['points'][i[0]] += 1

```
