```python
import pandas as pd
from pymongo import MongoClient
```

# То же самое но не из CSV а из БД

```python
client = MongoClient()
users_DB = client['VK_Inder']
users_collection = users_DB['users']
user_collection = users_DB['tihon']

df = pd.DataFrame(list(users_collection.find())).set_index('id')
df.head(1)
```

```python
df['points'] = 0
```

```python
user_df = pd.DataFrame(list(user_collection.find())).set_index('id')
user_df
```

```python
for i in user_df.T.iteritems():
    user_music_list = i[1][26].split(', ')
    user_films_list = i[1][25].split(', ')
    user_books_list = i[1][4].split(', ')
    user_games_list = i[1][15].split(', ')
```

```python
user_games_list
```

```python
#ищем общие книжки
for book in user_books_list:
    book = book.lower()
    for i in df.T.iteritems():
        if book in str(i[1][8]).lower():
            df.loc[df.index == i[0], 'points'] += 1
```

```python
#ищем общих друзей
for i in df.T.iteritems():
    if i[1][8] > 0:
        df.loc[df.index == i[0], 'points'] += 1
```

```python
#ищем общую музыку
for music in user_music_list:
    music = music.lower()
    for i in df.T.iteritems():
        if music in str(i[1][29]).lower():
            df.loc[df.index == i[0], 'points'] += 1
```

```python
#ищем общие фильмы
for film in user_films_list:
    film = film.lower()
    for i in df.T.iteritems():
        if film in str(i[1][6]).lower():
            df.loc[df.index == i[0], 'points'] += 1
```

```python
#ищем общие игры
for game in user_games_list:
    game = game.lower()
    for i in df.T.iteritems():
        if game in str(i[1][9]).lower():
            df.loc[df.index == i[0], 'points'] += 1
```

```python
top10 = df.sort_values(['points'], ascending=False).head(10)
```
